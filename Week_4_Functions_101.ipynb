## What is a Function?

In programming, a function is a reusable piece of code that performs a specific task. Functions help to make your code more readable, manageable, and efficient. Instead of writing the same code over and over, you can simply call the function.

## Defining a Function in Python

In Python, you define a function using the `def` keyword. Here's the syntax:

```python
def function_name(parameters):
    # code
    return result
```
- `function_name`: This is the name of the function. It should be descriptive of what the function does.
- `parameters`: These are the inputs to the function. A function can have any number of parameters, or none at all.
- `return`: This keyword returns the result of the function. The `return` statement is optional. If it's not included, the function will return `None`.

Here's an example of a Addition function:

```python
def add_numbers(x, y):
    result = x + y
    return result
```

Here's an example how to call the function:
```python
sum = add_numbers(3, 5)
print(sum)  # Outputs: 8
```

````python
import random

def coin_toss(toss_count):
    """
    Simulate a coin toss.
    """

    for toss in range(1,toss_count+1):
      choice = random.choice(["Heads", "Tails"])
      print(f"toss {toss} : {choice}")

    return None

coin_toss(toss_count=5)
````

## SNEAKY way of dealing with ERRORs

### TRY / EXCEPT

Errors are a way of life with technology. So knowing that errors (intentional or unintenntional) are going to occur in your Python program - why not deal with them?

When Python(or any language errors) it abruptly halts and spits out a message.

### The try and except Blocks
The try block contains code that might throw an exception. After the try block, include an except statement, followed by a block of code which handles the problem as elegantly as possible.

```python

try:
    # Code that might raise an exception
except ExceptionType:
    # Code that handles the exception

```



````python
try:
    # Code that might raise an exception
    x = 1 / 0
    print(x)
except:
    # Code to execute if an exception is raised
    print("An error occurred zero division is not allowed")
````

## How can i catch my Exception message?

easy ~ str(e)

````python
try:
    # Code that might raise an exception
    x = 1 / 0
except Exception as e:
    # Print the error message
    print("Caught an exception: ", str(e))
````

````python
# numeric variables
numeric_vars = poker.select_dtypes(include=['int64', 'float64']).columns.to_list()

numeric_vars
````

````python
# baseline code
column = 'hours_played_sitting'
summary_stat = poker.groupby('EVENT_LABEL')[column].describe().reset_index()
print(summary_stat)

plt.figure(figsize=(10,3))
sns.boxplot(poker,x='EVENT_LABEL',y=column)
plt.title(f"Boxplot of: {column}")
plt.xlabel(f"Fraud / Not Fraud")
plt.ylabel(f"{column}")
plt.show();
````

````python
# Wrapper this in a Function

def profile_number(column):
  """
    profile a numeric column.

    Parameters:
    column (int or float): column name to profile.

    Returns:
    None
  """

  summary_stat = poker.groupby('EVENT_LABEL')[column].describe().reset_index()
  print(f"-- stats for {column} -- ")
  print(summary_stat)

  plt.figure(figsize=(10,3))
  sns.boxplot(poker,x='EVENT_LABEL',y=column)
  plt.title(f"Boxplot of: {column}")
  plt.xlabel(f"Fraud / Not Fraud")
  plt.ylabel(f"{column}")
  plt.show();
  return None

column = 'hours_played_sitting'
profile_number(column)
````
